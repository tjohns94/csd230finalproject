```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
import sklearn as sk
```

```python
wwhs_en = pd.read_csv('./data/age_group_en.csv')
```

```python
# Assigning age_code to age_groups from 5 to 18 (used for linear data fitting)
age_code_dict = {
    '20 - 24': 5, '25 - 29': 6, '30 - 34': 7,
    '35 - 39': 8, '40 - 44': 9, '45 - 49': 10,
    '50 - 54': 11, '55 - 59': 12, '60 - 64': 13,
    '65 - 69': 14, '70 - 74': 15, '75 - 79': 16, 
    '80 - 84': 17, '85 - 89': 18
}
wwhs_en['age_code'] = wwhs_en['age_groups'].map(age_code_dict)
```

```python
# Visualize weight vs age (using age_code)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age_code', y='weight', data=wwhs_en)
plt.title('Weight vs Age')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```
